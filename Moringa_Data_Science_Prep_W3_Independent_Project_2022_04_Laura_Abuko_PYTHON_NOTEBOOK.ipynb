```python

#download,upload and loading our datatsets
import pandas as pd
cells_geo = pd.read_csv('cells_geo.csv')

#to print only a few rows use head inbuilt function
cells_geo.head()
#to print only a few rows use head inbuilt function

cells_geo = pd.read_csv('cells_geo.csv',delimiter=';', names = ['City','Status','Locality','Geographical_Zone','Zone_Name','Longitude','Latitude','Region','Area','Cell_ID','Site_ID'])
cells_geo.head()
```

```python
#changing column names for cellsgeodataset
cells_geo.columns = ['VILLES',	'STATUS',	'LOCALISATION' ,	'DECOUPZONE',	'ZONENAME',	'LONGITUDE',	'LATITUDE',	'REGION',	'AREA',	'CELL_ID', 'SITE_ID']
cells_geo.head()
```

```python
#loading excel file

cells_geo_description = pd.read_excel('cells_geo_description.xlsx')
cells_geo_description
```

```python
#loading last excel file

cdr_description = pd.read_excel('cdr_description.xlsx')
cdr_description
```

```python
#concatenating the twodatasets
description_combine = pd.concat([cells_geo_description,cdr_description,], ignore_index = 1)
description_combine.head()
```

```python
#Transposing our dataset
description_combine1 = description_combine.T
description_combine1
```

```python
#changing the first row to be the header
headers = description_combine1.iloc[0]
description_combine1 = pd.DataFrame(description_combine1.values[1:],columns=headers)
description_combine1
```

```python
#MERGING the twodatasets
pd.merge(cells_geo, description_combine1, on='LATITUDE')
```
